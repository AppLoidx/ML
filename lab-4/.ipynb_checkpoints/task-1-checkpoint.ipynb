## Создаем свой датасет

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.figsize']=(12,8)

from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=500, random_state=7, centers = 5)

plt.scatter(X[:,0], X[:,1])
```

```python
from sklearn.cluster import KMeans
kmeansModel = KMeans(n_clusters=2) # мы пока не знаем количество кластеров
```

```python
kmeansModel.fit(X) # обучаем модель
```

```python
labels = kmeansModel.labels_ # получаем метки
print(labels)
```

```python
plt.scatter(X[:,0], X[:,1], c=labels) # визуализируем данные
```

```python
criteries = []
for k in range(2,10):
  kmeansModel=KMeans(n_clusters=k, random_state=7)
  kmeansModel.fit(X)
  criteries.append(kmeansModel.inertia_)
```

```python
print(criteries)
```

```python
plt.plot(range(2,10), criteries)
```

как видно из графиков, оптимальная для нас точка это `k = 5`, после которого производная значения критерия ближе к нулю

```python
kmeansModel = KMeans(n_clusters=5) # используем k вычисленный через метод Локтя
kmeansModel.fit(X) # обучаем модель
plt.scatter(X[:,0], X[:,1], c=kmeansModel.labels_) # визуализируем данные
```

### Альтернативное решение через DBSCAN

```python
from sklearn.cluster import DBSCAN

clustering = DBSCAN(eps=1.55, min_samples=5).fit_predict(X)
print(clustering)
plt.scatter(X[:,0], X[:,1], c=clustering);
```
